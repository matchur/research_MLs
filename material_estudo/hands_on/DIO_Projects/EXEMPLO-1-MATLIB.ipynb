# EXEMPLO 1 - Regressão

Nesse exemplo vamos falar sobre uma técnica muito importante em Machine Learning, a **Regressão**.

Regressão é uma técnica utilizada para encontrar uma relação entre duas (ou mais) variáveis. Ou seja, ela tenta prever um valor numérico com base em outros dados.

Por exemplo, imagine que você tem um conjunto de dados com informações sobre uma variável de entrada **x** (como idade, temperatura ou altura), e quer prever um resultado **y** relacionado a ela (como salário, consumo de energia ou peso). A regressão tenta aprender essa relação observando exemplos passados.

A forma mais simples é a regressão linear, onde o modelo tenta ajustar uma reta que melhor representa os dados. A equação dessa reta seria algo como:

```bash
y = a * x + b
```

**Onde:**

- x é a variável de entrada (também chamada de feature),

- y é o valor que queremos prever (variável alvo),

- a é o coeficiente (inclinação da reta),

- b é o intercepto (ponto onde a reta cruza o eixo y).

## Codando

Vamos começar a parte da programação, ver isso na prática.

Começando com o import da biblioteca **matplotlib**, importante para visualização de dados.

````python
import matplotlib
````

Agora vamos importar do **scikit-learn.datasets** a função **make_regression**

O código a seguir gera um conjunto de 200 amostras (n_samples=200), com apenas 1 variável de entrada (n_features=1) e um nível de variação ou ruído (noise=30) nos dados. O ruído simula variações naturais que podem ocorrer no mundo real.

O resultado é um par de variáveis:

**x**: os valores de entrada (ou preditores),

**y**: os valores de saída (ou alvos), calculados com base em uma relação linear com x, mas com alguma variação aleatória.

A partir desses dados, é possível treinar um modelo de regressão para prever novos valores de **y** a partir de **x**.

````python
from sklearn.datasets import make_regression
x , y = make_regression(n_samples=200,n_features=1,noise=30)

````

Agora vamos gerar o gráfico com matplotlib para visualizar os dados de X,Y.

````python
plt.scatter(x, y)
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Dados gerados para regressão")
plt.show()
````

Agora que temos alguns dados, vamos treinar um modelo.

É nessa parte que começa a "aprendizagem" de verdade.

````python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(x, y)     
````

Agora que já geramos os dados e treinamos o modelo com .fit(x, y),
podemos usar o método .predict(x) para fazer previsões.

O .predict() serve para aplicar a fórmula que o modelo aprendeu (do tipo y = a * x + b)
e gerar valores estimados de y, com base nos valores de x que passamos.

Em outras palavras, o modelo agora já "entende" a relação entre x e y,
então conseguimos dar novos valores de x e ele vai nos dizer qual y ele espera.

Isso é o que chamamos de previsão — é o objetivo principal de usar um modelo de regressão.

````python
y_pred = modelo.predict(x) 
````

Aqui estamos criando novos valores de entrada (x) para testar o modelo.

Esses valores representam casos onde ainda não sabemos o y real,
e queremos que o modelo nos dê uma estimativa com base no que aprendeu.

Usamos np.array para criar os valores em formato de matriz, como o modelo espera.
Em seguida, usamos .predict() para gerar as previsões de y para esses novos x. (50, 100, 150)

````python
import numpy as np

novos_x = np.array([[50], [100], [150]])
novas_predicoes = modelo.predict(novos_x)
````

````python
import matplotlib.pyplot as plt

plt.scatter(x, y, color='blue', label='Dados reais')
plt.plot(x, y_pred, color='red', label='Reta predita pelo modelo')
plt.legend()
plt.title("Regressão linear sobre dados simulados")
plt.show()
````
