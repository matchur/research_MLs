# Exercício 1 - Redução de Dimensionalidade
Aqui está um exercício simples de visão computacional, focando em redução de dimensionalidade de imagens com Python, usando o OpenCV e Matplotlib.

Vamos usar uma imagem em RGB, convertê-la para tons de cinza e, em seguida, para uma imagem binária. 
Também aplicaremos alguns filtros e malhas para explorar a transformação da imagem.

## Passo 1: Importar as bibliotecas

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Passo 2: Carregar a imagem e exibir as informações
Aqui, vamos carregar uma imagem e mostrar algumas informações sobre ela, como o formato e o tipo.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Verifique se o arquivo de imagem existe
image_path = 'imagem_teste.png'  # Substitua pelo caminho correto da sua imagem
if not os.path.exists(image_path):
    print("A imagem não foi encontrada no caminho especificado.")
else:
    # Carregar a imagem
    img = cv2.imread(image_path)

    # Verificar se a imagem foi carregada corretamente
    if img is None:
        print("Erro ao carregar a imagem. Verifique o formato ou o caminho.")
    else:
        # Exibir informações sobre a imagem
        print("Tipo da imagem:", type(img))
        print("Dimensões da imagem:", img.shape)
        print("Imagem em RGB:", img)

        # Mostrar a imagem original
        plt.figure(figsize=(10, 5))
        plt.subplot(131)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title('Imagem Original')
        plt.axis('off')
```

## Passo 3: Converter para tons de cinza
Agora, vamos converter a imagem para tons de cinza, que é uma forma de redução de dimensionalidade, já que estamos eliminando a informação das cores.

```python
# Converter para tons de cinza
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Exibir a imagem em tons de cinza
plt.subplot(132)
plt.imshow(gray_img, cmap='gray')
plt.title('Imagem em Tons de Cinza')
plt.axis('off')
```

## Passo 4: Converter para imagem binária (binarização)
A binarização é uma técnica de redução de dimensionalidade, onde convertemos a imagem para apenas duas cores (preto e branco), usando um limiar.

```python
# Aplicar binarização
_, binary_img = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)

# Exibir a imagem binária
plt.subplot(133)
plt.imshow(binary_img, cmap='gray')
plt.title('Imagem Binária')
plt.axis('off')
```

## Passo 5: Aplicar filtro (Exemplo de filtro de desfoque)
Agora, vamos aplicar um filtro simples (desfoque) na imagem binária.

```python
# Aplicar filtro de desfoque na imagem binária
blurred_img = cv2.GaussianBlur(binary_img, (5, 5), 0)

# Exibir a imagem borrada
plt.figure(figsize=(5, 5))
plt.imshow(blurred_img, cmap='gray')
plt.title('Imagem Borrada')
plt.axis('off')
```

## Passo 6: Aplicar malhas (Exemplo de malha de bordas)
Para explorar mais transformações, podemos aplicar um detector de bordas como o filtro de Canny, para destacar as bordas da imagem.

```python
# Aplicar o detector de bordas de Canny
edges = cv2.Canny(binary_img, 100, 200)

# Exibir as bordas
plt.figure(figsize=(5, 5))
plt.imshow(edges, cmap='gray')
plt.title('Bordas da Imagem')
plt.axis('off')
plt.show()
```
